```python
import pickle
import re
from transformers import (Trainer, TrainingArguments, AutoTokenizer, HfArgumentParser,
                          DataCollatorForLanguageModeling)
```

```python
with open("./dataset_store/train_dataset", "rb") as f:
    train_dataset = pickle.load(f)
```

```python
tokenizer = AutoTokenizer.from_pretrained('gpt2')
tokenizer.model_max_length = 64
tokenizer.pad_token = tokenizer.eos_token
```

```python
tokenizer.eos_token
```

```python
def transform(sentence):

    return re.sub('[^A-Za-z0-9.,;!?]+', ' ', sentence) + "<|endoftext|>"
```

```python
for i in range(50):
    sentence_transf = transform(train_dataset[i]['text'])
    print(f"--- Raw : {sentence_transf}")
    x = tokenizer.encode(sentence_transf)
    print(f"--- encode : {x}")
    print(f"--- and decode : {tokenizer.decode(x)}")
    print("----")
```
